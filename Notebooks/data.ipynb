```python
from faker import Faker
import random
import csv

fake = Faker()

# Predefined skill pool
skill_pool = [
    "Python", "JavaScript", "TypeScript", "C++", "C#", "Java", "Go", "Ruby", "PHP", "Rust",
    "SQL", "NoSQL", "Django", "Flask", "FastAPI", "Spring Boot", "React", "Vue.js", "Angular",
    "Node.js", "Next.js", "MongoDB", "PostgreSQL", "MySQL", "SQLite", "HTML", "CSS", "SASS",
    "Tailwind", "Bootstrap", "Material UI", "Redux", "GraphQL", "REST API", "gRPC",
    "Scikit-Learn", "Pandas", "Numpy", "Matplotlib", "Seaborn", "TensorFlow", "PyTorch",
    "Transformers", "LLMs", "Hugging Face", "OpenCV", "Keras", "XGBoost", "CatBoost",
    "Kubernetes", "Docker", "AWS", "Azure", "GCP", "Firebase", "Jenkins", "Terraform",
    "Git", "GitHub Actions", "CI/CD", "Linux", "Bash", "PowerShell", "Nginx", "Apache",
    "ElasticSearch", "Redis", "RabbitMQ", "Kafka", "Matlab", "Tableau", "Power BI", "Figma",
    "JIRA", "Notion", "Airflow", "Spark", "Hadoop", "BigQuery", "LangChain"
]


# Role type and domain word banks
roles_types = [
    "Engineer", "Developer", "Architect", "Scientist", "Specialist", "Strategist",
    "Consultant", "Analyst", "Manager", "Lead", "Intern", "Instructor", "Administrator",
    "Coach", "Evangelist"
]

domains = [
    "AI", "Data", "Cloud", "Frontend", "Backend", "DevOps", "ML", "NLP", "Cybersecurity",
    "Blockchain", "Analytics", "Web3", "Mobile", "AR/VR", "Quantum", "IoT", "Embedded",
    "Product", "Site Reliability", "Platform", "Game", "UI/UX", "Automation", "Research",
    "Education", "Testing", "Security", "CRM", "E-commerce"
]

# Generate unique roles
def generate_unique_roles(num_roles=100):
    generated_roles = {}
    used_names = set()

    while len(generated_roles) < num_roles:
        domain = random.choice(domains)
        role_type = random.choice(roles_types)
        role_name = f"{domain} {role_type}"

        if role_name not in used_names:
            used_names.add(role_name)
            num_skills = random.randint(4, 7)
            selected_skills = random.sample(skill_pool, num_skills)
            generated_roles[role_name] = selected_skills

    return generated_roles

# Write to CSV
def write_roles_to_csv(roles_dict, filename="generated_roles.csv"):
    with open(filename, mode='w', newline='', encoding='utf-8') as file:
        writer = csv.writer(file)
        writer.writerow(["Role", "Skills"])  # Header

        for role, skills in roles_dict.items():
            skills_str = ", ".join(skills)
            writer.writerow([role, skills_str])

if __name__ == "__main__":
    roles = generate_unique_roles()
    write_roles_to_csv(roles)
    print("âœ… Role data saved to 'generated_roles.csv'")
```
